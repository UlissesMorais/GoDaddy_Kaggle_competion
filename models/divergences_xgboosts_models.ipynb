```python
blacklistcfips_baseline_dic = dict(zip(blacklistcfips_baseline, [1]*len(blacklistcfips_baseline)))
blacklistcfips_modified_dic = dict(zip(blacklistcfips_modified, [1]*len(blacklistcfips_modified)))
```

```python
def calc_diff(df1, df2):
    return 2 * 100 * ((df1 - df2).abs() / (df1 + df2).abs())
```

```python
# merge both submission sets
sub_baseline = sub_baseline.rename(columns={'microbusiness_density': 'mbd_baseline'})
sub_modified = sub_modified.rename(columns={'microbusiness_density': 'mbd_modified'})
sub_merged = sub_baseline.merge(sub_modified[['row_id','mbd_modified']], on='row_id', how='inner')

# mapping
sub_merged['blklist_baseline'] = sub_merged['cfips'].map(blacklistcfips_baseline_dic).fillna(0)
sub_merged['blklist_modified'] = sub_merged['cfips'].map(blacklistcfips_modified_dic).fillna(0)
sub_merged['sum_blk'] = sub_merged['blklist_modified'] + sub_merged['blklist_baseline']

# simple average ensemble
sub_merged['microbusiness_density'] = (sub_merged['mbd_baseline'] + sub_merged['mbd_modified']) / 2

# calculate biggest diferences in predictions
sub_merged['pct_diff'] = calc_diff(sub_merged['mbd_baseline'], sub_merged['mbd_modified'])

sub_merged.head()
```
